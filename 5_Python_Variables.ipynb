# **Python Variables**

## What is a Variable?

Variables are containers used to store data values.

A variable gives a name to a piece of data so that it can be used and modified later in the program.

## Creating Variables

Python has no separate command for declaring variables.

A variable is created automatically when you assign a value to it for the first time.

```python
x = 5
y = "John"

print(x)
print(y)
```

## Dynamic Typing in Python

Python is a dynamically typed language.

This means:

- You do not need to specify the data type while creating a variable.
- Python automatically determines the data type based on the assigned value.

## Variables Can Change Their Type

A variable can hold different types of values during program execution.

```python
x = 4       # x is an integer
x = "Sally" # x is now a string

print(x)
```

## Casting

Casting means converting a value from one data type to another.

Python provides built-in functions for type conversion:

- `str()` → Converts a value to a string
- `int()` → Converts a value to an integer
- `float()` → Converts a value to a floating-point number

```python
x = str(3)    # x will be '3'
y = int(3)    # y will be 3
z = float(3)  # z will be 3.0

print(x)
print(y)
print(z)
```

## Understanding the Data Types

| Variable | Value | Data Type |
|----------|--------|-----------|
| `x` | `'3'` | `str` |
| `y` | `3` | `int` |
| `z` | `3.0` | `float` |

## Key Takeaways

- Variables are containers for storing data values.
- Variables are created when a value is assigned to them.
- Python does not require variable declarations.
- Python automatically determines the data type of variables.
- A variable can change from one data type to another.
- Casting is used to explicitly convert values into a specific data type.
- Common casting functions are `str()`, `int()`, and `float()`.
