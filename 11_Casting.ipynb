# Python Casting

## What is Casting?

Sometimes, you may want to convert a value from one data type to another.

This process is called **casting** or **type conversion**.

Python is an object-oriented language, and data type conversion is performed using constructor functions.

## Common Constructor Functions for Casting

Python provides the following built-in functions for type conversion:

- `int()` → Converts a value to an integer.
- `float()` → Converts a value to a floating-point number.
- `str()` → Converts a value to a string.

# Integer Casting (`int()`)

## The `int()` Function

The `int()` function creates an integer value from:

- An integer
- A floating-point number (by removing the decimal part)
- A string containing a whole number

```python
x = int(1)
y = int(2.8)
z = int("3")

print(x)
print(y)
print(z)
```

### Important Note

`int()` does **not round** decimal numbers.

It simply removes the decimal part (truncation).

```python
print(int(35.88))
print(int(9.99))
print(int(-7.8))
```

# Float Casting (`float()`)

## The `float()` Function

The `float()` function creates a floating-point number from:

- An integer
- A float
- A string containing an integer or decimal value

```python
x = float(1)
y = float(2.8)
z = float("3")
w = float("4.2")

print(x)
print(y)
print(z)
print(w)
```

# String Casting (`str()`)

## The `str()` Function

The `str()` function converts values into string representations.

It can convert:

- Strings
- Integers
- Floating-point numbers
- Many other data types

```python
x = str("s1")
y = str(2)
z = str(3.0)

print(x)
print(y)
print(z)
```

## Checking the Resulting Data Types

```python
x = int("10")
y = float(10)
z = str(10)

print(type(x))
print(type(y))
print(type(z))
```

## Invalid Type Conversion

Casting fails when the value cannot be converted into the requested type.

```python
int("Hello")
```

# Key Takeaways

- Casting means converting one data type into another.
- Python uses constructor functions for type conversion.
- `int()` converts values to integers by removing the decimal part.
- `float()` converts values to floating-point numbers.
- `str()` converts values into strings.
- Use `type()` to verify the resulting data type after conversion.
- Invalid conversions raise exceptions such as `ValueError`.
