### Python Strings

#### What are Strings?
A **string** is a sequence of characters used to represent text.
In Python, strings are enclosed within:
* Single quotes (' ')
* Double quotes (" ")

#### Creating Strings

````
print("Hello")
print('Hello')
````

##### Output

```text
Hello
Hello
```

#### Quotes Inside Quotes
You can use quotation marks inside a string as long as they do not match the surrounding quotes.

````
print("It's alright")
print("He is called 'Johnny'")
print('He is called "Johnny"')
````

##### Output

```text
It's alright
He is called 'Johnny'
He is called "Johnny"
```

#### Assigning a String to a Variable
Strings can be stored in variables using the assignment operator (=).

````
a = "Hello"
print(a)
````

##### Output

```text
Hello
```

### Multiline Strings

#### Creating Multiline Strings
Use three double quotes (""") or three single quotes (''') to create strings that span multiple lines.

````
a = """Lorem ipsum dolor sit amet, consectetur adipiscing elit, sed do eiusmod tempor incididunt ut labore et dolore magna aliqua."""
print(a)
````

````
a = '''Lorem ipsum dolor sit amet, consectetur adipiscing elit, sed do eiusmod tempor incididunt ut labore et dolore magna aliqua.'''
print(a)
````

##### Note
The line breaks in the output appear exactly as they are written in the code.

### Strings are Arrays

#### Strings as Sequences of Characters
Strings in Python are sequences (arrays) of Unicode characters.
Python does not have a separate character data type. A single character is simply a string of length 1.
Characters can be accessed using their index positions.

````
a = "Hello, World!"
print(a[1])
````

##### Output

```text
e
```

##### Index Positions

### Looping Through a String

Since strings are sequences, we can iterate through each character using a for loop.

````
for x in "banana":
    print(x)
````

##### Output

```text
b
a
n
a
n
a
```

### String Length

#### The len() Function
Use the len() function to determine the number of characters in a string.

````
a = "Hello, World!"
print(len(a))
````

##### Output

```text
13
```

### Checking if a String Contains Text

#### The in Keyword
Use the in keyword to determine whether a substring exists inside a string.

````
txt = "The best things in life are free!"
print("free" in txt)
````

##### Output

```text
True
```

#### Using in with an if Statement

````
txt = "The best things in life are free!"
if "free" in txt:
    print("Yes, 'free' is present.")
````

##### Output

```text
Yes, 'free' is present.
```

### Checking if a String Does NOT Contain Text

#### The not in Keyword
Use the not in keyword to determine whether a substring is absent from a string.

````
txt = "The best things in life are free!"
print("expensive" not in txt)
````

##### Output

```text
True
```

#### Using not in with an if Statement

````
txt = "The best things in life are free!"
if "expensive" not in txt:
    print("No, 'expensive' is NOT present.")
````

##### Output

```text
No, 'expensive' is NOT present.
```

### Key Takeaways
* Strings are sequences of characters enclosed in single or double quotes.
* Triple quotes allow the creation of multiline strings.
* Strings are immutable sequences of Unicode characters.
* Characters are accessed using indexes starting from 0.
* Strings can be iterated using a for loop.
* Use len() to determine the length of a string.
* Use in to check if text exists in a string.
* Use not in to check if text does not exist in a string.
