### Python - Slicing Strings

#### What is Slicing?
**Slicing**  allows you to extract a portion (substring) of a string.
Python uses the following syntax:
*  start → Index where slicing begins (included)
*  end → Index where slicing stops (not included)
**Note:**  The first character has an index of 0.

#### Basic Slicing
Specify the starting and ending positions separated by a colon (:).
Syntax:

````python
b = "Hello, World!"
print(b[2:5])
````

##### Output
```text
llo
```

#### Understanding the Index Positions
b[2:5]
*  Start at index 2
*  Stop before index 5
*  Characters at indexes 2, 3, and 4 are returned.

### Slice From the Start

If the starting index is omitted, Python starts from index 0.
Syntax:

````python
b = "Hello, World!"
print(b[:5])
````

##### Output
```text
Hello
```

### Slice To the End
If the ending index is omitted, Python continues until the last character.
Syntax:

````python
b = "Hello, World!"
print(b[2:])
````

##### Output
```text
llo, World!
```

### Negative Indexing
Python supports  **negative indexing** , where counting starts from the end of the string.

````python
b = "Hello, World!"
print(b[-5:-2])
````

##### Output
```text
orl
```

### More Examples
The same slicing rules can be applied to any string.

````python
text = "Python"
print(text[0:2])
print(text[:4])
print(text[2:])
print(text[-3:])
````

##### Output
```text
Py
Pyth
thon
hon
```

### Key Takeaways
*  Slicing extracts a portion of a string.
*  Syntax: string[start:end]
*  The starting index is included.
*  The ending index is excluded.
*  Omitting start begins from index 0.
*  Omitting end continues to the last character.
*  Negative indexing counts from the end of the string.
*  Slicing returns a new string and does not modify the original string.
