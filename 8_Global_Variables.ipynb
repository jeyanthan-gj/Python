# Global Variables

## What are Global Variables?

Variables that are created outside a function are called **global variables**.

Global variables can be accessed:

- Outside the function
- Inside the function

## Using a Global Variable Inside a Function

```python
x = "awesome"

def myfunc():
    print("Python is " + x)

myfunc()
```

## Local Variables and Global Variables

A variable created inside a function is called a **local variable**.

A local variable:

- Exists only inside the function.
- Cannot be accessed outside the function.
- Can have the same name as a global variable.

```python
## Local Variable with the Same Name as a Global Variable

x = "awesome"

def myfunc():
    x = "fantastic"
    print("Python is " + x)

myfunc()

print("Python is " + x)
```

### Explanation

- The variable `x = "fantastic"` is local to `myfunc()`.
- It exists only inside the function.
- The global variable `x = "awesome"` remains unchanged.

# The `global` Keyword

## Creating a Global Variable Inside a Function

Normally, variables created inside a function are local.

To create a global variable inside a function, use the `global` keyword.

```python
def myfunc():
    global x
    x = "fantastic"

myfunc()

print("Python is " + x)
```

## Modifying a Global Variable Inside a Function

To change the value of an existing global variable inside a function, use the `global` keyword.

```python
x = "awesome"

def myfunc():
    global x
    x = "fantastic"

myfunc()

print("Python is " + x)
```

# Key Takeaways

- Variables created outside functions are called **global variables**.
- Global variables can be accessed both inside and outside functions.
- Variables created inside functions are called **local variables**.
- A local variable can have the same name as a global variable.
- Use the `global` keyword to create a global variable inside a function.
- Use the `global` keyword to modify an existing global variable inside a function.
- Without `global`, changes inside a function affect only the local variable.
