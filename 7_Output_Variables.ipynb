# Output Variables

## Output Variables

The `print()` function is commonly used to display values and variables on the screen.

```python
## Printing a Single Variable

x = "Python is awesome"
print(x)
```

## Printing Multiple Variables Using Commas

You can output multiple variables by separating them with commas inside the `print()` function.

```python
x = "Python"
y = "is"
z = "awesome"

print(x, y, z)
```

## Printing Multiple Variables Using the `+` Operator

You can also combine strings using the `+` operator.

```python
x = "Python "
y = "is "
z = "awesome"

print(x + y + z)
```

### Note

Notice the spaces after `"Python "` and `"is "`.

Without these spaces, the output would be:

`Pythonisawesome`

## Using `+` with Numbers

For numeric values, the `+` operator performs mathematical addition.

```python
x = 5
y = 10

print(x + y)
```

## Mixing Strings and Numbers with `+`

The `+` operator cannot directly combine a string and a number. Doing so results in a `TypeError`.

```python
x = 5
y = "John"

print(x + y)
```

## Recommended Way: Use Commas

The best way to output values of different data types is to separate them with commas in the `print()` function.

```python
x = 5
y = "John"

print(x, y)
```

## Key Takeaways

- Use `print()` to display variables and values.
- Multiple variables can be printed by separating them with commas.
- The `+` operator can concatenate strings.
- The `+` operator performs addition when used with numbers.
- Strings and numbers cannot be directly combined using `+`.
- Using commas in `print()` is the recommended way to display values of different data types.
