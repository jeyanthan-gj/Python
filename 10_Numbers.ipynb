# Python Numbers

## Numeric Data Types in Python

Python has three built-in numeric data types:

1. `int`
2. `float`
3. `complex`

## Creating Numeric Variables

Variables of numeric types are created when values are assigned to them.

```python
x = 1      # int
y = 2.8    # float
z = 1j     # complex
```

## Checking the Data Type

Use the `type()` function to determine the data type of a variable.

```python
x = 1
y = 2.8
z = 1j

print(type(x))
print(type(y))
print(type(z))
```

# Integer (`int`)

## What is an Integer?

An integer (`int`) is a whole number without a decimal point.

Integers can be:

- Positive
- Negative
- Very large numbers

```python
x = 1
y = 35656222554887711
z = -3255522

print(type(x))
print(type(y))
print(type(z))
```

# Floating Point Numbers (`float`)

## What is a Float?

A float is a number that contains one or more decimal points.

Floats can be:

- Positive
- Negative
- Decimal numbers

```python
x = 1.10
y = 1.0
z = -35.59

print(type(x))
print(type(y))
print(type(z))
```

## Scientific Notation

Floats can also be represented using scientific notation with `e` or `E`, which represents powers of 10.

```python
x = 35e3
y = 12E4
z = -87.7e100

print(x)
print(y)
print(z)

print(type(x))
print(type(y))
print(type(z))
```

# Complex Numbers (`complex`)

## What is a Complex Number?

Complex numbers contain a real part and an imaginary part.

In Python, the imaginary part is represented using `j`.

```python
x = 3 + 5j
y = 5j
z = -5j

print(type(x))
print(type(y))
print(type(z))
```

# Type Conversion

## Converting Between Numeric Types

Python provides built-in functions for converting numeric values:

- `int()`
- `float()`
- `complex()`

```python
x = 1      # int
y = 2.8    # float
z = 1j     # complex

a = float(x)
b = int(y)
c = complex(x)

print(a)
print(b)
print(c)

print(type(a))
print(type(b))
print(type(c))
```

## Important Note

Complex numbers cannot be converted into `int` or `float`.

The following code will generate an error:

```python
x = 3 + 5j
int(x)
```

# Random Numbers

## Generating Random Numbers

Python does not have a built-in `random()` function.

Instead, Python provides a module called `random` for generating random numbers.

```python
import random

print(random.randrange(1, 10))
```

**Note:** The output can be any integer from `1` to `9`. Each execution may produce a different result.

# Key Takeaways

- Python has three numeric data types: `int`, `float`, and `complex`.
- `int` represents whole numbers.
- `float` represents decimal and scientific notation numbers.
- `complex` represents numbers with an imaginary part using `j`.
- Use `type()` to determine the data type of a variable.
- Use `int()`, `float()`, and `complex()` for numeric type conversion.
- Complex numbers cannot be converted to `int` or `float`.
- Random numbers can be generated using the `random` module.
