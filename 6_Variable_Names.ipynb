## What are Variable Names?

A variable name is the identifier used to refer to a variable in a program.

Variable names can be short, such as `x` and `y`, or descriptive, such as `age`, `car_name`, and `total_volume`.

## Rules for Naming Variables in Python

Python variable names must follow these rules:

1. A variable name must start with a letter (`a-z`, `A-Z`) or an underscore (`_`).
2. A variable name cannot start with a number.
3. A variable name can contain only letters, numbers, and underscores (`_`).
4. Variable names are case-sensitive.
5. A variable name cannot be a Python keyword.

## Legal Variable Names

The following are valid variable names in Python:

```python
myvar = "John"
my_var = "John"
_my_var = "John"
myVar = "John"
MYVAR = "John"
myvar2 = "John"
```

## Illegal Variable Names

The following variable names are invalid:

```python
# SyntaxError

2myvar = "John"   # Cannot start with a number
my-var = "John"   # Hyphens are not allowed
my var = "John"   # Spaces are not allowed
```

## Variable Names are Case-Sensitive

Python treats uppercase and lowercase letters differently.

Therefore, `age`, `Age`, and `AGE` are three different variables.

```python
age = 20
Age = 25
AGE = 30

print(age)
print(Age)
print(AGE)
```

# One Value to Multiple Variables

## Assigning the Same Value to Multiple Variables

Python allows you to assign the same value to multiple variables in a single line.

```python
x = y = z = "Orange"

print(x)
print(y)
print(z)
```

## Key Takeaways

- Variable names identify and store data in memory.
- Variable names must follow Python's naming rules.
- Variable names are case-sensitive.
- Multiple variables can be assigned the same value in one statement.
