# **Python Statements**

## What is a Statement?

A computer program is a list of instructions that are executed by a computer.

In programming languages, these instructions are called **statements**.

A statement tells Python to perform a specific task.

## Example of a Statement

The following statement prints the text `"Python is fun!"` on the screen:

````python
print("Python is fun!")
````

## Statements Usually End with a New Line

In Python, a statement normally ends when the line ends.

Unlike languages such as C, C++, or Java, Python does **not** require a semicolon (`;`) at the end of every statement.

## Multiple Statements

Most Python programs contain many statements.

Python executes statements one by one in the same order in which they are written.

````python
print("Hello World!")
print("Have a good day.")
print("Learning Python is fun!")
````

## How Python Executes These Statements

1. First statement executes:
   ```python
   print("Hello World!")
   ```

2. Second statement executes:
   ```python
   print("Have a good day.")
   ```

3. Third statement executes:
   ```python
   print("Learning Python is fun!")
   ```

Python executes statements from **top to bottom**, one statement at a time.

## Semicolons in Python

Semicolons (`;`) are optional in Python.

You can place multiple statements on the same line by separating them with semicolons. However, this is rarely used because it reduces code readability.

````python
print("Hello")
print("How are you?")
print("Bye bye!")
````

### Same Code Using Semicolons

````python
print("Hello"); print("How are you?"); print("Bye bye!")
````

## Invalid Syntax

If two statements are written on the same line without a newline or semicolon separator, Python generates a `SyntaxError`.

````python
print("Python is fun!") print("Really!")
````

## Key Takeaways

- A **statement** is an instruction given to Python.
- Python executes statements **one by one** in the order they are written.
- Most statements end automatically when the line ends.
- Semicolons (`;`) are optional.
- Multiple statements can be written on one line using semicolons, but it is not recommended.
- Writing multiple statements on one line without a separator causes a `SyntaxError`.
