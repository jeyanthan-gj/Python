## Executing Python Code

Python code can be executed in two ways:

1. Directly in the Python Interpreter (Command Line)
2. By creating and running a Python file (`.py`)

## Method 1: Using the Python Interpreter

You can type Python statements directly into the Python interpreter.

**Example:**

````python
print("Hello, World!")
````

## Method 2: Using a Python File

You can write Python code in a file with the `.py` extension and execute it from the command line.

For example, create a file named `myfile.py` and run it using:




```
C:\Users\Your Name>python myfile.py
```



# Python Indentation

**Indentation** refers to the spaces at the beginning of a line of code.

In many programming languages, indentation is used only to improve readability. However, in Python, indentation is **mandatory** because it defines blocks of code.

## Why is Indentation Important?

Python uses indentation to indicate a block of code, such as:

- Conditions (`if`, `else`)
- Loops (`for`, `while`)
- Functions
- Classes

## Correct Indentation Example

````python
if 5 > 2:
    print("Five is greater than two!")
````

## Missing Indentation Causes an Error

If you do not indent the code inside a block, Python raises an error.

````python
# Syntax Error

if 5 > 2:
print("Five is greater than two!")
````

## How Many Spaces Should Be Used?

The number of spaces is up to the programmer. However:

- At least one space is required.
- The most commonly used indentation is **4 spaces**.

````python
if 5 > 2:
 print("Five is greater than two!")
````

````python
if 5 > 2:
        print("Five is greater than two!")
````

## Consistent Indentation is Required

All statements inside the same block must use the same level of indentation. Mixing different numbers of spaces in the same block results in an error.

````python
# Syntax Error

if 5 > 2:
    print("Five is greater than two!")
        print("Five is greater than two!")
````

## Key Takeaways

- Python uses indentation to define code blocks.
- Indentation is mandatory in Python.
- Four spaces are the recommended standard.
- All statements within the same block must have consistent indentation.
- Incorrect indentation results in an `IndentationError`.
